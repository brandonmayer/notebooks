### Defined radius from Target Point (1000km)

Created a radius of 1000km from our target point to create an area to look for all known tracks which have intersected this zone

1000km on the Mercator projection is roughly 9 degrees (both lat/lon) here, so had a quick look at max wind speed as a funtion of distance

````python
Image('max_intensity_distance_degrees.png')
````

Max speed did not increase beyond this 1000km radius and it includes only the tracks that move west of Madagascar (see next plot) at some point. 70 cyclones were found.

````python
Image('002-cm9_buffer.png')
````

Shown next are the 70 tracks which intersected (at some point) the defined zone (in gray)

Note: track data often only included time, longitude and latitude. Central pressure and max winds were not included in a significant part of the sample

````python
Image('003-tracks.png')
````

Using all the points associated with this set of tracks, the probability density function (PDF) was estimated with a kernel density estimation (shown below).

This establishes a surface where the probability of any storm (known) to have occurred in the region might track. No intensity data was used in this estimation.

````python
Image('base-kde.png')
````

With this PDF surface (note the PDF was simply inverted to represent minimums), a series of random minimizations (from points distributed over area) using the gradient in a truncated Newton algorithm were computed, with each step of the minimzation kept track of to determine the path to minimization found for various starting points (1000 random points were used)

The result highlights the average track historical storms indicate cyclones in this region are most likely to follow

````python
Image('opt-paths.png')
````

I was unable to automate the line generation from this data, but shown below are two tracks that use the above information, hand drawn in this case, both following the same(similar) path once west of Madagascar. The red track originates from north, the green further east in the central Indian ocean

````python
Image('opt_tracks.png')
````

Now that two likely tracks have been identified using historical track data, assigning realistic (but extreme) values for pressure and winds needed to be done.

As mentioned before, there was much more track data than pressure/wind data.

Because we're dealing with extremes, I had a look at the longitudinal and latitudunal distribution of MAX wind speed (knots), MIN central pressure (mbar) and computed forward speed(knots), binned by 0.25 degree bands.

e.g. the 40E-40.25E longitudinal band shows the minimum pressure (mbar) found amongst all storms to have a report in that band.

What is seen is a fairly realistic representation of storm progression longitudinally and latitudinally.

Using the two defined tracks, the max wind, min central pressure, and max forward speed between the longitudinal bin and latitudinal bin, was recorded at the storms intensity at that point.
