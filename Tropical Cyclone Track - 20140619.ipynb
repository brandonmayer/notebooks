Iterate through all the known tracks in Southern Indian Ocean.

If any point along track intersects the buffer defined by radius from point, include.

```python
def haversine(lat1, lon1, lat2, lon2):
 
  R = 6372.8 # Earth radius in kilometers
 
  dLat = math.radians(lat2 - lat1)
  dLon = math.radians(lon2 - lon1)
  lat1 = math.radians(lat1)
  lat2 = math.radians(lat2)
 
  a = math.sin(dLat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dLon/2)**2
  c = 2*math.asin(math.sqrt(a))
 
  return R * c
```

```python
points = []
tracks = []
llwpf = []
with open('tracks.json', 'r') as f:
    storms = json.load(f)
    for s in storms.keys():
        if all(key in storms[s] for key in ['lat','lon','pressure','wind','time']):
            # check length of each component is the same
            if len(set([len(x) for x in storms[s].values()])) != 1:
                print 'unequal lengths of keyed arrays, eg. len(lat) != len(lon)'
                continue
            time = storms[s]['time']
            lat = storms[s]['lat']
            lon = storms[s]['lon']
            prs = storms[s]['pressure']
            wnd = storms[s]['wind']

            fwd = [0.0] # filler value so all arrays same len
            # compute forward speed from lonlat and time (result will be one smaller than above arrays)
            for x in xrange(1, len(time)):
                delta_t = (np.datetime64(time[x]) - np.datetime64(time[x-1])).item().total_seconds()
                d = haversine(lat[x-1],lon[x-1],lat[x],lon[x])
                #d = 300 * a.distance(b) * math.cos(math.radians((lat[x-1]+lat[x])/2.0)) #TODO this for haversine?               
                fwd.append(d/delta_t*1943.84449) #km/s to knots
            
            if len(lon) > 2 and len(lat) > 2: # need at least one point with
                # skip junk data
                if all([w == 0.0 for w in wnd]):
                    continue
                if all([p == 0.0 for p in prs]):
                    continue
                track = LineString(zip(lon,lat))
                # only consider if intersects the buffer
                if track.intersects(CM9_buffer):
                    points.append((lon,lat))
                    tracks.append(track)
                    llwpf.append((zip(lon,lat,wnd,prs,fwd))) # REMEMBER TO DISREGARD THE FIRST VALUE!
```

```python
# kernel density estimate of the PDF (DO THIS IN LATLON!!)
kde = gaussian_kde([tx,ty])

# bounds
bounds = (np.amin(tx),np.amin(ty),np.amax(tx),np.amax(ty))

# evaluate the estimated PDF on a grid
x,y = np.mgrid[bounds[0]:bounds[2]:0.25,bounds[1]:bounds[3]:0.25]
z = kde((x.ravel(),y.ravel())).reshape(*x.shape)

# plot
pc = m.pcolor(x,y,z,latlon=True,alpha=0.5,zorder=3)
plt.savefig('base-kde.png')
Image('base-kde.png')
```

```python
"""
invert PDF surface and find minimums from random spots, saving the 'path' along the way to minimum
"""
def cost(x):
    """ searching for minimum, invert PDF """
    return kde(x) * -1.0

points = [] # path points along all optimizations ('like rolling marbles')
def steps_callback(xk):
    """ creates array of points, one for each step in the optimization """
    global steps
    steps.append(tuple(xk))
    
n = 500 # number of random points to start from
starts = zip(np.random.uniform(bounds[0],bounds[2],n),np.random.uniform(bounds[1],bounds[3],n))
for i, s in enumerate(starts):
    steps = [] #resets the globally 'steps' used in def callback()
    o = fmin_tnc(cost, s, approx_grad=1, epsilon=1e-08, disp=0, maxfun=1000, callback=steps_callback)
    points.extend(steps) # this just adds the steps of this optimization to the points, loses association of a path

tx,ty = zip(*points)

# WKT MultiPoint (used in GIS tool to draw tracks)
mp_f = open('track.wkt', 'w')
mp = MultiPoint([Point(*xy) for xy in points])
mp_f.write(mp.wkt)
mp_f.close()
```
